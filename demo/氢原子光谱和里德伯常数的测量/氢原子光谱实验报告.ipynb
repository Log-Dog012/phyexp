## 1. 用钠黄光 λ = 589.3 nm 作为标准谱线校准光栅常数 d，并计算不确定度 u(d)。注意消除正负级不严格对称的系统误差。

正负级不严格对称的系统误差的消除：

通过正负级做差除二作为折射角测量结果消除系统误差

```python
钠灯角度["k=±2差值 左"]=钠灯角度["k=-2 左"] - 钠灯角度["k=+2 左"]
钠灯角度["k=±2差值 右"]=钠灯角度["k=-2 右"] - 钠灯角度["k=+2 右"]
钠灯角度["k=±2差值"]=(钠灯角度["k=±2差值 右"]+钠灯角度["k=±2差值 左"])/2
钠灯角度["k=±2差值弧度"]=钠灯角度["k=±2差值"].apply(lambda x: x.to('rad').magnitude)  # 转为弧度数值
钠灯角度["k=±2差值弧度"]=钠灯角度["k=±2差值弧度"].apply(lambda x: x+np.pi if x < 0 else x)  # 调整到0-pi区间
钠灯角度["折射角"]=钠灯角度["k=±2差值弧度"]/2
```

```python
钠灯角度
```

*最后两列被转换为弧度制，下面的计算也按弧度制进行*

```python
theta=钠灯角度["折射角"].mean()
```

```python
u_atheta=钠灯角度["折射角"].std()
```

```python
仪器误差限=分度值=ureg("1arcmin").to("rad").magnitude
u_btheta=仪器误差限/3**0.5
```

```python
utheta=np.sqrt(u_atheta**2+u_btheta**2)
```

```python
theta=ufloat(theta,utheta)
```

```python
theta
```

$$d \sin \theta=k \lambda$$

```python
k=2
lam=ureg("589.3nm")
d=k*lam/umath.sin(theta)
```

```python
d
```

$$u(d)=\frac{k \lambda \cos \theta}{\sin^2 \theta} u(\theta)$$

上面的结果是自动微分算出的，与使用公式的结果相同

```python
ud=(2*lam*np.cos(theta.n)/np.sin(theta.n)**2*theta.s).magnitude
ud
```

## 2. 应用不等精度加权平均法计算氢原子的里德伯常数的最佳值 ，并与标准值比较，得出相对偏差。

```python
氢灯折射角=pd.DataFrame({
    "k=1":(氢灯角度["k=-1 左"]+氢灯角度["k=-1 右"]-氢灯角度["k=+1 左"]-氢灯角度["k=+1 右"])/4,
    "k=2":(氢灯角度["k=-2 左"]+氢灯角度["k=-2 右"]-氢灯角度["k=+2 左"]-氢灯角度["k=+2 右"])/4,
    "k=3":(氢灯角度["k=-3 左"]+氢灯角度["k=-3 右"]-氢灯角度["k=+3 左"]-氢灯角度["k=+3 右"])/4,
})
```

```python
氢灯折射角.index=[3,4,5]
```

```python
氢灯折射角
```

```python
for i in 氢灯折射角:
    氢灯折射角[i]=氢灯折射角[i].apply(lambda x: x.to('rad').magnitude)
    氢灯折射角[i]=氢灯折射角[i].apply(lambda x: x+np.pi/2 if x < 0 else x)
```

```python
氢灯折射角
```

$$d \sin \theta=k \lambda$$

$$\frac{1}{\lambda}=R_H \left( \frac{1}{2^2}-\frac{1}{n^2} \right)$$

由上两式

$$R_H=\frac{k}{d \sin{\theta} \times (\frac{1}{2^2}-\frac{1}{n^2})}$$

为简便，记$f(d,\theta)=\frac{1}{d \sin{\theta}}$

$$u(R_H)=\frac{k}{\frac{1}{2^2}-\frac{1}{n^2}} \sqrt{ \left( \frac{\partial f}{\partial d} \right)^2 u^2(d) + \left( \frac{\partial f}{\partial \theta} \right)^2 u^2(\theta)}$$

$$\frac{\partial f}{\partial d}=-\frac{1}{d^2 \sin \theta}$$
$$\frac{\partial f}{\partial \theta}=-\frac{\cos \theta}{d \sin^2 \theta}$$

由于每组只测量了一次，所以

$$u(\theta)=u_b(\theta)$$

后续的数值计算是通过自动微分实现的，应当与公式结果一致

```python
def 计算里德伯常数(k,n,theta):
    theta=ufloat(theta,u_btheta)
    lam=d*umath.sin(theta)/k
    RH=1/lam/(2**-2-n**-2)
    return RH
```

```python
里德伯常数=[]
for i in 氢灯折射角:
    for j in range(3,6):
        result=计算里德伯常数(int(i[-1]),j,氢灯折射角[i][j])
        if not np.isnan(result.n):
            里德伯常数.append(result)
```

```python
里德伯常数
```

```python
R_H最佳观测量=sum(i.n/i.s**2 for i in 里德伯常数)/sum(1/i.s**2 for i in 里德伯常数)
```

```python
R_H最佳观测量
```

```python
uR_H=np.sqrt(1/sum(1/i.s**2 for i in 里德伯常数))
```

```python
uR_H
```

```python
R_H=ufloat(R_H最佳观测量,uR_H)/ureg("nanometer")
```

```python
R_H.to("1/m")
```

测量得到里德伯常数为$(1.0890 \pm 0.0005) \times 10^7 \  {meter}^{-1}$

```python
from scipy.constants import physical_constants

# 获取里德伯常数的详细信息
value, unit, uncertainty = physical_constants['Rydberg constant']

R_H_0=quan(value,unit)
```

```python
R_H_0
```

```python
相对偏差=((R_H最佳观测量/ureg("nanometer")-R_H_0)/R_H_0).to("").magnitude
print(f"{相对偏差=:.3%}")
```
